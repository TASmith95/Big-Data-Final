# Question 6

```
#Question 6
from pyspark.sql import SparkSession
from pyspark.ml.clustering import KMeans
from pyspark.ml.feature import VectorAssembler,StringIndexer
from pyspark.ml import Pipeline

data=df.select(['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds'])

```

```
data=data.dropna()
```

```


MntWines_indexer = StringIndexer(inputCol='MntWines',outputCol='MntWines_index',handleInvalid='keep')
MntFruits_indexer = StringIndexer(inputCol='MntFruits',outputCol='MntFruits_index',handleInvalid='keep')
MntMeatProducts_indexer = StringIndexer(inputCol='MntMeatProducts',outputCol='MntMeatProducts_index',handleInvalid='keep')
MntFishProducts_indexer = StringIndexer(inputCol='MntFishProducts',outputCol='MntFishProducts_index',handleInvalid='keep')
MntSweetProducts_indexer = StringIndexer(inputCol='MntSweetProducts',outputCol='MntSweetProducts_index',handleInvalid='keep')
MntGoldProds_indexer = StringIndexer(inputCol='MntGoldProds',outputCol='MntGoldProds_index',handleInvalid='keep')
```

```
assembler = VectorAssembler(inputCols=['MntWines_index','MntFruits_index','MntMeatProducts_index',
                                       'MntFishProducts_index','MntSweetProducts_index','MntGoldProds_index'],
                            outputCol="features")
```

```
# Pipeline is used to pass the data through indexer and assembler simultaneously. Also, it helps to pre-rocess the test data
# in the same way as that of the train data.
pipe = Pipeline(stages=[MntWines_indexer,MntFruits_indexer,MntMeatProducts_indexer,MntFishProducts_indexer,MntSweetProducts_indexer,MntGoldProds_indexer,assembler])
```

```
final_data=pipe.fit(data).transform(data)
```

```
# Create an object for the KMeans model

kmeans_model = KMeans(k=6)
```

```
fit_model = kmeans_model.fit(final_data)
```

```
wssse = fit_model.summary.trainingCost 
print("The within set sum of squared error of the mode is {}".format(wssse))
```

```
centers = fit_model.clusterCenters()
```

```
print("Cluster Centers")
index=1
for cluster in centers:
    print("Centroid {}: {}".format(index,cluster))
    index+=1
```

```
# Store the results in a dataframe

results = fit_model.transform(final_data)
```

```
results.head()
```

```
results.select(['MntWines_index','MntFruits_index','MntMeatProducts_index','MntFishProducts_index','MntSweetProducts_index','MntGoldProds_index']).show()
```

```
results.groupby('prediction').count().sort('prediction').show()
```

```
from pyspark.ml.evaluation import ClusteringEvaluator
from pyspark.ml.clustering import KMeans
silhouette_score=[]
evaluator = ClusteringEvaluator(predictionCol='prediction', featuresCol='features', \
                                metricName='silhouette', distanceMeasure='squaredEuclidean')
for i in range(2,10):
    
    KMeans_algo=KMeans(k=i)
    KMeans_fit=KMeans_algo.fit(final_data)
    output=KMeans_fit.transform(final_data)
    score=evaluator.evaluate(output)
    silhouette_score.append(score)
    print("Silhouette Score:",score)
```

```
#Visualizing the silhouette scores in a plot
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,1, figsize =(8,6))
ax.plot(range(2,10),silhouette_score)
ax.set_xlabel('k')
ax.set_ylabel('cost')
```
